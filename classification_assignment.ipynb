```python
df = pd.read_csv("CKD.csv")
```

```python
df = pd.get_dummies(df, columns = ["cad","appet","pe","sg","rbc","pc","pcc","ba","htn","dm","ane"], drop_first=True)    
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
col_features = ["age","bp","al","su","bgr","bu","sc","sod","pot","hrmo","pcv","wc","rc"]
```

```python
df[col_features] = scaler.fit_transform(df[col_features])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
independant = df.drop("classification", axis = 1)
```

```python
independant.info()
```

```python
dependant = df["classification"]
```

```python
dependant.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    independant, dependant, 
    test_size=0.2,      # 20% of data goes to the test set
    random_state=42,    # Ensures you get the same split every time
    stratify=dependant          # Recommended: maintains the same ratio of classes as the original data
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier

```

```python
model1 = LogisticRegression()
model2 = GaussianNB()
model3 = DecisionTreeClassifier()
model4 = LinearSVC()
model5 = KNeighborsClassifier()
model1.fit(X_train, y_train)
```

```python
model2.fit(X_train, y_train)
```

```python
model3.fit(X_train, y_train)
```

```python
model4.fit(X_train, y_train)
```

```python
model5.fit(X_train, y_train)
```

```python
y1_pred = model1.predict(X_test)
```

```python
y2_pred = model2.predict(X_test)
```

```python
y3_pred = model3.predict(X_test)
```

```python
y4_pred = model4.predict(X_test)
```

```python
y5_pred = model5.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix 
cm1 = confusion_matrix(y_test, y1_pred)
print(classification_report(y_test, y1_pred))
```

```python
cm2 = confusion_matrix(y_test, y2_pred)
print(classification_report(y_test, y2_pred))
```
